```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-100, 200, 4)
y = np.arange(-100, 200, 4)
X, Y = np.meshgrid(x, y)
```

```python
def f6(x, y):
    f6_1 = np.sin(np.sqrt(x ** 2 + y ** 2)) ** 2 - 0.5
    f6_2 = (1 + 0.001 * (x ** 2 + y ** 2)) ** 2
    ans = 0.5 + (f6_1 / f6_2)
    return ans
```

```python
ANS = f6(X, Y)
```

```python
plt.figure(figsize=(30,20))
ax=plt.subplot(1,1,1,projection="3d")
ax.plot_surface(X,Y,ANS,rstride=1,cstride=1,alpha=0.3,cmap=plt.get_cmap('rainbow'))
ax.set_zticks((0,0.3))
ax.view_init(15,50)
plt.show()
```

```python
for i in range(0, 180,1):
    plt.figure(figsize=(15,10))
    ax=plt.subplot(1,1,1,projection="3d")
    ax.plot_surface(X,Y,ANS,rstride=1,cstride=1,alpha=0.3,cmap=plt.get_cmap('rainbow'))
    ax.set_zticks((0,0.3))
    ax.view_init(15,i)
    plt.savefig("function6_{0}.png".format(i))
```

```python
mylist=[]
for arry in range(0,180):
    mylist.append("function6_{0}.png".format(arry))
```

```python
import imageio  

def create_gif(image_list, gif_name):  
    frames = []  
    for image_name in image_list:  
        frames.append(imageio.imread(image_name))  
    # Save them as frames into a gif   
    imageio.mimsave(gif_name, frames, 'GIF', duration = 0.1)  
    return  
  
def main():  
    image_list = mylist
    gif_name = 'created_gif.gif'  
    create_gif(image_list,"function.gif")  
main()
```
